```python
#Exercise 1: Simple Polynomial Plot
import numpy as np
import matplotlib.pyplot as plt

# Données
x = np.linspace(-10, 10, 100)
y = 2*x**3 - 5*x**2 + 3*x - 7

# Tracé
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b')  # ligne bleue
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tracé du polynôme 2x³ - 5x² + 3x - 7')
plt.grid(True)
plt.show()
```

```python
# Exercise 2: Exponential and Logarithmic Plot

# Données
x = np.linspace(0.1, 10, 500)
y1 = np.exp(x)
y2 = np.log(x)

# Tracé
plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'r--', label='exp(x)')   # ligne rouge en pointillés
plt.plot(x, y2, 'g-', label='log(x)')    # ligne verte pleine
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fonctions exponentielle et logarithmique')
plt.grid(True)
plt.legend()
plt.savefig('exponentiel_logarithme.png', dpi=100)
plt.show()
```

```python
#Exercise 3: Figure and Subplots

import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-2*np.pi, 2*np.pi, 400)
x2 = np.linspace(-2, 2, 400)

# Création des sous-graphes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Premier subplot
ax1.plot(x1, np.tan(x1), label='tan(x)')
ax1.plot(x1, np.arctan(x1), label='arctan(x)')
ax1.set_title('tan(x) et arctan(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

# Deuxième subplot
ax2.plot(x2, np.sinh(x2), label='sinh(x)')
ax2.plot(x2, np.cosh(x2), label='cosh(x)')
ax2.set_title('sinh(x) et cosh(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
# EXO 4

n = np.random.randn(1000)  # valeurs normales (loi normale)

plt.figure(figsize=(8, 5))
plt.hist(n, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogramme des valeurs normales')
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.xlim([min(n), max(n)])
plt.grid(True)
plt.show()
```

```python
# Exercise 5: Scatter Plot

x = np.random.uniform(0, 10, 500)
y = np.sin(x) + np.random.normal(0, 0.2, 500)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, c=y, s=50, cmap='viridis', alpha=0.7)
plt.title('Scatter plot de y = sin(x) + bruit')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xticks([])
plt.yticks([])
plt.savefig('scatter_plot.pdf')
plt.show()
```

```python
# Exercise 6: Contour Plot

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure()
contour = plt.contourf(X, Y, Z, cmap='plasma')  # remplissage coloré
plt.colorbar()
plt.title("Tracé de contours de sin(sqrt(x² + y²))")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Exercise 7: 3D Surface Plot

from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5.25, 0.25)
Y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf)
ax.set_title("Surface 3D de cos(sqrt(X² + Y²))")
plt.show()
```

```python
# Exercise 8: Line and Marker Styles
x = np.linspace(-2, 2, 10)
y1 = x**2
y2 = x**3
y3 = x**4

plt.plot(x, y1, 'r--o', label='x^2')  # rouge, tirets, ronds
plt.plot(x, y2, 'g-^', label='x^3')   # vert, plein, triangles
plt.plot(x, y3, 'b-.s', label='x^4')  # bleu, pointillés, carrés
plt.title("Différents styles de courbes")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise 9: Logarithmic Scale
x = np.logspace(0, 2, 50)  # de 10^0 à 10^2
y1 = 2**x
y2 = np.log2(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y1, label='2^x')
plt.plot(x, y2, label='log2(x)')
plt.yscale('log')  # échelle logarithmique pour l'axe y
plt.title("Tracés avec échelle logarithmique")
plt.xlabel("x")
plt.ylabel("y (échelle log)")
plt.grid(True, which='both')
plt.legend()
plt.show()
```

```python
#Exercise 10: Changing Viewing Angle
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5.25, 0.25)
Y = np.arange(-5, 5.25, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(12, 5))

# Première vue
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='coolwarm')
ax1.view_init(30, 45)
ax1.set_title("Vue angle 30°, 45°")

# Deuxième vue
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='coolwarm')
ax2.view_init(60, 120)
ax2.set_title("Vue angle 60°, 120°")

plt.tight_layout()
plt.show()
```

```python
# Exercise 11: 3D Wireframe Plot

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Création des grilles X et Y allant de -5 à 5 avec un pas de 0.25
X, Y = np.meshgrid(np.arange(-5, 5.25, 0.25), np.arange(-5, 5.25, 0.25))

# Calcul de la fonction Z = sin(X) * cos(Y)
Z = np.sin(X) * np.cos(Y)

# Création de la figure 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé en wireframe (maillage) de Z
ax.plot_wireframe(X, Y, Z, color='black')

# Changement de l'angle de vue pour une meilleure perspective
ax.view_init(45, 30)

# Titre du graphe
ax.set_title("Tracé Wireframe de sin(X) * cos(Y)")

plt.show()
```

```python
# Exercise 12: 3D Contour Plot

# Création des grilles X et Y allant de -5 à 5 avec un pas de 0.25
X, Y = np.meshgrid(np.arange(-5, 5.25, 0.25), np.arange(-5, 5.25, 0.25))

# Calcul de Z = exp(-0.1 * (X² + Y²)), une fonction en forme de cloche
Z = np.exp(-0.1 * (X**2 + Y**2))

# Création de la figure 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé des courbes de niveau en 3D (contour3D)
ax.contour3D(X, Y, Z, 50, cmap='inferno')  # 50 niveaux de contour, palette de couleurs inferno

# Ajout des titres et étiquettes
ax.set_title("Contour 3D de exp(-0.1*(X² + Y²))")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()

```

```python
#Exercise 13: 3D Parametric Plot
# Création d'un tableau t de 100 points entre 0 et 2π
t = np.linspace(0, 2*np.pi, 100)

# Définition des coordonnées paramétriques : hélice
X = np.sin(t)
Y = np.cos(t)
Z = t

# Création de la figure 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé de la courbe paramétrique en 3D
ax.plot3D(X, Y, Z, color='purple')

# Ajout des titres et étiquettes
ax.set_title("Courbe paramétrique 3D : X = sin(t), Y = cos(t), Z = t")
ax.set_xlabel("X = sin(t)")
ax.set_ylabel("Y = cos(t)")
ax.set_zlabel("Z = t")

plt.show()
```

```python
# Exercise 14: 3D Bar Plot

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Création des points X et Y (10 valeurs entre -5 et 5)
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Calcul de Z = exp(-0.1*(X² + Y²))
Z = np.exp(-0.1 * (X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Convertir les matrices en vecteurs pour le bar3D
x = X.ravel()
y = Y.ravel()
z = np.zeros_like(x)
dz = Z.ravel()

# Dimensions fixes pour chaque barre
dx = dy = 0.4

# Tracé 3D en barres
ax.bar3d(x, y, z, dx, dy, dz, cmap='viridis', shade=True)

# Étiquettes
ax.set_title("Barres 3D de exp(-0.1*(x² + y²))")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
# Exercise 15: 3D Vector Field
# Grilles 3D de -5 à 5
X, Y, Z = np.meshgrid(np.arange(-5, 6, 2),
                      np.arange(-5, 6, 2),
                      np.arange(-5, 6, 2))

# Champ vectoriel : U, V, W
U = -Y  # direction X
V = X   # direction Y
W = Z   # direction Z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé du champ vectoriel avec quiver3D
ax.quiver(X, Y, Z, U, V, W, color='blue', length=1.5)

# Étiquettes
ax.set_title("Champ vectoriel 3D")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
# Exercise 16: 3D Scatter Plot

# Génération de 100 points aléatoires pour x, y, z selon une loi normale
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter 3D avec couleurs selon z
sc = ax.scatter(x, y, z, c=z, cmap='plasma')
fig.colorbar(sc)

# Étiquettes
ax.set_title("Nuage de points 3D coloré par z")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
# Exercise 17: 3D Line Plot

# Courbe paramétrique : t de 0 à 4π
t = np.linspace(0, 4*np.pi, 100)
X = np.sin(t)
Y = np.cos(t)
Z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé de la ligne 3D
ax.plot3D(X, Y, Z, color='green', linewidth=2)

# Étiquettes
ax.set_title("Ligne 3D : sin(t), cos(t), t")
ax.set_xlabel("X = sin(t)")
ax.set_ylabel("Y = cos(t)")
ax.set_zlabel("Z = t")

plt.show()
```

```python
# Exercise 18: 3D Filled Contour Plot

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Création des grilles X et Y
x = np.arange(-5, 5.1, 0.1)
y = np.arange(-5, 5.1, 0.1)
X, Y = np.meshgrid(x, y)

# Calcul de Z = sin(sqrt(X² + Y²))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Création de la figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracé du contour rempli en 3D
cont = ax.contourf(X, Y, Z, 50, cmap='viridis')  # 50 niveaux, avec remplissage

# Barre de couleurs
fig.colorbar(cont)

# Titres et axes
ax.set_title("Contour rempli 3D de sin(sqrt(X² + Y²))")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
# Exercise 19: 3D Heatmap

# Génération d'une grille de 50x50 points
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calcul de Z = sin(sqrt(x² + y²))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Affichage avec imshow (heatmap 2D pour représenter Z)
plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', cmap='plasma')
plt.colorbar(label='Valeurs de Z')
plt.title("Heatmap de sin(sqrt(x² + y²))")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
